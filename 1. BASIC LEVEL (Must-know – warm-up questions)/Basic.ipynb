# 1. Write a Python program to reverse a string without using built-in reverse functions.


```python
lst = [10,20,30]
rev_lst = lst[::-1]
print(rev_lst)
```

# 2. Check whether a given number is a palindrome.

```python
num = input("Enter a Number: ")
rev_num = num[::-1]
if num == rev_num:
    print("palindrome")
else:
    print("Not a Palindrome")
```

# 6.Check whether a number is prime.


```python
num = int(input("Enter a Number: "))

if num<=1:
    print("Not a Prime")
else:
    for i in range(2,num):
        if num%i==0:
            print("Not a Prime")
            break
        else:
            print("Prime")
            break
```

# 7. Print the Fibonacci series up to n terms.

```python
num = int(input("Enter a Number: "))

a =0
b =1
for i in range(num):
    print(a)
    a,b=b,a+b
```

# 8. Find the sum of digits of a given number.


```python
num = int(input("Enter a Number: "))
sum = 0
while num>0:
    last_digit = num%10
    sum+=last_digit
    num = num //10

print(sum)
```

# 9. Remove duplicate elements from a list without using set().


```python
lst = [3,6,1,8,3,0,7,2,7,3,6]
rem_duplicat = []

for num in lst:
    if num not in rem_duplicat:
        rem_duplicat.append(num)
print(sorted(rem_duplicat))
```

# 10. Find the second largest element in a list. 

```python
lst = [10,20,30,40,50]
largest = second_largest = float('-inf')

for num in lst:
    if num > largest:
        second_largest = largest
        largest = num
    elif num > second_largest or num !=largest:
        second_largest = num
print(second_largest)
```
