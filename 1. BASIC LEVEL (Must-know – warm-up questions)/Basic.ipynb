### 2. Check whether a given number is a palindrome.


```python
num = int(input("Enter a number: "))

original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

if original == reverse:
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

### 3. Find the largest and smallest number in a list without using max() and min().


```python
def max_min(numbers):
    minimum=maximun= numbers[0]
    for i in numbers:
        if i<minimum:
           minimum = i
        elif i>maximun:
            maximun = i  
    return minimum, maximun

numbers = [3,6,2,79,1,53,2,8,54]

print(max_min(numbers))
```

```python
def max_mini(lst):
    maximum = max(lst)
    minimun = min(lst)
    return maximum, minimun

numbers = [3,6,2,79,1,53,2,8,54]
max_mini(numbers)
```

### 4.Count the frequency of each character in a string.


```python
string = "Python is a powerfull language"
count = {}

for ch in string:
    if ch in count:
        count[ch]+=1
    else :
        count[ch]=1

print(count)
```

### 6. Check whether a number is prime.


```python
def is_prime(n):
    if n <= 1:
        return f"Not a Prime No."
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return f"Not a Prime No."
    return f"Prime No."
    

num = int(input("enter a number: "))
print(is_prime(num))
```

### 7. Print the Fibonacci series up to n terms.


```python
def fibonacci(n):
    a, b = 0, 1
    series = []
    for _ in range(n):
        series.append(a)
        a,b=b,a+b
    return series

n = int(input("Enter Your No.:"))
print(fibonacci(n))
```

### 8. Find the sum of digits of a given number.


```python
def sum_of_digit(n):
    sum =0 
    while n>0:
        sum += n%10
        n/=10
    return sum
n = 423
print(sum_of_digit(n))
```

### 9. Remove duplicate elements from a list without using set().


```python
def duplicate_element(lst):
    duplicat= []
    for i in lst:
        if i not in duplicat:
            duplicat.append(i)
    return duplicat

lst = [1,2,3,4,4,4,5,6,2,2,7,7] 
print(duplicate_element(lst))
```

### 10. Find the second largest element in a list.


```python
def second_largest(lst):
    first = second = float('-inf')
    for i in lst:
        if i>first:
            second=first
            first=i
        elif first>i>second:
            first=second
    return second

lst = [10,20,40,41,70,30]
print(second_largest(lst))
```
