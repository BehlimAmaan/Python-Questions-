### 2. Check whether a given number is a palindrome.


```python
num = int(input("Enter a number: "))

original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

if original == reverse:
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

### 3. Find the largest and smallest number in a list without using max() and min().


```python
def max_min(numbers):
    minimum=maximun= numbers[0]
    for i in numbers:
        if i<minimum:
           minimum = i
        elif i>maximun:
            maximun = i  
    return minimum, maximun

numbers = [3,6,2,79,1,53,2,8,54]

print(max_min(numbers))
```

```python
def max_mini(lst):
    maximum = max(lst)
    minimun = min(lst)
    return maximum, minimun

numbers = [3,6,2,79,1,53,2,8,54]
max_mini(numbers)
```

### 4.Count the frequency of each character in a string.


```python
string = "Python is a powerfull language"
count = {}

for ch in string:
    if ch in count:
        count[ch]+=1
    else :
        count[ch]=1

print(count)
```

### 6. Check whether a number is prime.


```python
def is_prime(num):
    if num<=1:
        print("Not a prime number")
    else :
        for i in range(2, num):
            if num % i ==0:
                print("Not a prime No.")
            else:
                print("Its a prime No.")
    return num
    

num = int(input("enter a number: "))
print(is_prime(num))
```
