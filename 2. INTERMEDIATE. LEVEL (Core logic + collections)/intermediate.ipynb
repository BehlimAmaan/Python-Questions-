# 1. Check whether two strings are anagrams.


```python
str1 = "listen"
str2 = 'silent'

if sorted(str1) == sorted(str2):
    print("Anagrams")
else:
    print("Not Anagrams")
```

# 2. Find all duplicate elements in a list.

```python
lst = [10,20,30,10,20,40,20,50,10]
duplicate = []

for i in lst:
    if lst.count(i) > 1 and i not in duplicate:
        duplicate.append(i) 
    

print(duplicate)
```

# 4. Merge two dictionaries. If a key is common, add their values.


```python
dict1 = {"a": 20000, 'b':50000}
dict2 = {"a": 40000,'b':10000,'c':20000}

result = dict1.copy()

for key, value in dict2.items():
    if key in result:
        result[key]+=value
    else:
        result[key]=value
print(result)
```

# 5. Rotate a list by k positions.


```python
lst = [1,2,3,4,5]
k = int(input("Enter Value of K: "))

k = k % len(lst)
rotated = lst[-k:] + lst[:-k]
print(rotated)
```

# 6. Find missing numbers in a list containing numbers from 1 to n.


```python

def missing(lst):
    missing_number = []
    for num in range(1, max(lst)):
        if num not in lst:
            missing_number.append(num)
    return missing_number
lst = [1,2,4,5,6,9]
print(missing(lst))
```

# 7. Check if a list is sorted or not.


```python

def check_sorted(lst):
    for num in range(len(lst)-1):
        if lst[num] > lst[num+1]:
            return "Not Sorted"
        else:
            return "yes Sorted"
lst = [2,1,2,3,4,5,6]
print(check_sorted(lst))
```

# 8. Reverse each word in a sentence without changing word order.


```python
sentence = "Machine Learning is best. Machine Learning is very Powerfull"

sentence_split = sentence.split()
def rev_word(sentence):
    rev_word = []
    for i in sentence_split:
        rev_word.append(i[::-1])
        rev = " ".join(rev_word)
    return rev
print(f"Original String: {sentence}\nTransformed: {rev_word(sentence)}")
```
