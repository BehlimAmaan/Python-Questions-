### 1. Check whether two strings are anagrams.


```python
def anagrams(lst1, lst2):
    if sorted(lst1) == sorted(lst2):
        return f"Anagrams: {lst1} & {lst2}"
    else:
        return f"Not Anagrams: {lst1} & {lst2}"

# Example usage
print(anagrams("listen", "silent"))  # Anagrams
print(anagrams("hello", "world"))    # Not Anagrams
```
